Handling TV Show Genre column data

```python
listGenres=set(["Action Genre","Animation","Comedy Genre","Crime","Drama","Experimental","Fantasy","Historical","Horror","Romance","Science Fiction","Thriller","Western","Anime"])
for i in df.index:
  try:
    temp=df["genres"][i]
    temp=temp.split(",")
    temp=[j.strip() for j in temp]
    for k in range(len(temp)):
      if temp[k] not in listGenres:
        temp[k]="other TVShow Genre"
    df["genres"][i]=temp
  except:
    pass
listGenres.add("otherTVShowGenre")
```

```python
for i in listGenres:
  temp_list=[]
  for j in df.index:
    try:
      if i in df["genres"][j]:
        temp_list.append(1)
      else:
        temp_list.append(0)
    except:
      pass
  df[i]=temp_list
```

```python
df.drop(labels="genres",axis=1,inplace=True)
df.head()
```

TV Genre Preferences


```python
# Let's start by visualizing the distribution of gender in the dataset.  
fig, ax = plt.subplots()  
# Counting the Genres of TV Shows and Movies for each person and summing them up
genreCount=[]
for i in listGenres:
  genreCount.append(df[i].sum())
def sort_list(list1, list2):
    zipped_pairs = zip(list2, list1)
    l1 = [y for x, y in sorted(zipped_pairs)]
    return l1
listGenres=sort_list(listGenres,genreCount)
genreCount.sort()
ax.bar(listGenres, genreCount)  
ax.set_xlabel('TV Show Genre')  
ax.set_ylabel('Count')  
plt.xticks(rotation=90)
plt.show()  
```

Handling Music Genre column data

```python
listMusicGenres=set(["Blues","Jazz","Rock Music","Country","Pop","Hip Hop","EDM","K-POP","Rap"])
for i in df.index:
  try:
    temp=df["music"][i]
    temp=temp.split(",")
    temp=[j.strip() for j in temp]
    for k in range(len(temp)):
      if temp[k] not in listMusicGenres:
        temp[k]="otherMusic"
    df["music"][i]=temp
  except:
    pass
listMusicGenres.add("otherMusic")
```

```python
for i in listMusicGenres:
  temp_list=[]
  for j in df.index:
    try:
      if i in df["music"][j]:
        temp_list.append(1)
      else:
        temp_list.append(0)
    except:
      pass
  df[i]=temp_list
```

```python
df.drop(labels="music",axis=1,inplace=True)
df.head()
```

# CHECKING COLLINEARITY

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(30,30))     
sns.heatmap(df.corr(), annot=True, linewidths=.5, ax=ax)
```

When variables used in clustering have strong correlation, some variables get a higher weight than others. If two variables are perfectly correlated, they effectively represent the same concept. But that concept is now represented twice in the data and hence gets twice the weight of all the other variables. The final solution is likely to be skewed in the direction of that concept, which could be a problem if it’s not anticipated

# STRONG CORRELATION FEATURES LIST -

```python
corr = df.corr()
c1 = corr.abs()
upper = c1.where(np.triu(np.ones(c1.shape), k=1).astype(np.bool))
# Find index of feature columns with correlation greater than 0.5
to_drop = [column for column in upper.columns if any(upper[column] >= 0.5)]
print(to_drop)
```

#  CLUSTERING ALGORITHMS

### HIERARCHICAL CLUSTERING

```python
import scipy.cluster.hierarchy as hc
linkage_ward=hc.linkage(newdf, method='ward',optimal_ordering=True)
linkage_complete=hc.linkage(newdf, method='complete',optimal_ordering=True)
linkage_centroid=hc.linkage(newdf, method='centroid',optimal_ordering=True)
linkage_average=hc.linkage(newdf, method='average',optimal_ordering=True)
linkage_weighted=hc.linkage(newdf, method='weighted',optimal_ordering=True)
```

```python
from scipy.spatial.distance import pdist
for i in [linkage_ward,linkage_complete,linkage_centroid,linkage_average,linkage_weighted]:
    c, coph_dists = hc.cophenet(i, pdist(newdf))
    print(c)
```

```python
linkage=linkage_centroid
```

```python
plt.figure(figsize=(20, 10))  
plt.title("Dendrograms")  
dend = hc.dendrogram(linkage)
plt.axhline(y=14, color='r', linestyle='--')
```

```python
from sklearn import metrics
from sklearn.cluster import KMeans
for i in range(2,7):
  fl = hc.fcluster(linkage,i,criterion='maxclust')
  print(metrics.silhouette_score(newdf, fl, metric='euclidean'))
```

# PERFORMANCE EVALUATION METRICS

```python
from sklearn import metrics

cluster_metrics=pd.DataFrame(columns=['silhoutte score','db score','ch score'],index=['complete hierarchical','kmeans','dbscan'])
cluster_metrics['silhoutte score']['complete hierarchical']=metrics.silhouette_score(newdf, fl, metric='euclidean')
cluster_metrics['silhoutte score']['kmeans']=metrics.silhouette_score(newdf, y_kmeans, metric='euclidean')
cluster_metrics['silhoutte score']['dbscan']=metrics.silhouette_score(newdf, db_labels, metric='euclidean')

cluster_metrics['db score']['complete hierarchical']=metrics.davies_bouldin_score(newdf, fl)
cluster_metrics['db score']['kmeans']=metrics.davies_bouldin_score(newdf, y_kmeans)
cluster_metrics['db score']['dbscan']=metrics.davies_bouldin_score(newdf, db_labels)

cluster_metrics['ch score']['complete hierarchical']=metrics.calinski_harabasz_score(newdf, fl)
cluster_metrics['ch score']['kmeans']=metrics.calinski_harabasz_score(newdf, y_kmeans)
cluster_metrics['ch score']['dbscan']=metrics.calinski_harabasz_score(newdf, db_labels)
```

```python
cluster_metrics
```

Other evaluation metrics are not usable since this is an unsupervised clustering.

Silhoutte score and CH score for k-means is a little better than "complete" hierarchical clustering
